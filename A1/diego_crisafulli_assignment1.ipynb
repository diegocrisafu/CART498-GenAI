```python
from diffusers import StableDiffusionPipeline
import torch
import os
from transformers import CLIPImageProcessor  # Import for CLIP (though not directly used)

# Suppress the specific FutureWarning about CLIPFeatureExtractor
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="transformers.models.clip.feature_extraction_clip")

# Set the device (GPU if available, otherwise CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained(
    "Lykon/dreamshaper-8",
    torch_dtype=torch.float16,
    variant="fp16"
).to(device)

# --- Define paths relative to your Google Drive ---
# Assuming "diego_crisafulli_assignment1" is in the root of your Google Drive
project_root = "/content/drive/MyDrive/diego_crisafulli_assignment1"
prompts_dir = os.path.join(project_root, "prompts")
images_dir = os.path.join(project_root, "images")

# --- Create the images directory if it doesn't exist (prompts should already exist) ---
if not os.path.exists(images_dir):
    os.makedirs(images_dir)

# --- Generate and save images ---
for i in range(1, 4):
    prompt_filename = f"prompt_{i:02d}.txt"
    prompt_path = os.path.join(prompts_dir, prompt_filename)

    # Check if the prompt file exists
    if not os.path.exists(prompt_path):
        print(f"Error: Prompt file not found: {prompt_path}")
        print("Make sure you have uploaded the prompt files to your Google Drive!")
        continue  # Skip to the next iteration

    # Read the prompt from the file
    with open(prompt_path, "r") as f:
        prompt = f.read().strip()

    # Generate the image
    image = pipe(prompt, num_inference_steps=50).images[0]

    # Create a filename for the image
    image_filename = f"collage_{i:02d}.jpg"
    image_path = os.path.join(images_dir, image_filename)

    # Save the image
    image.save(image_path)
    print(f"Image saved to {image_path}")
```
